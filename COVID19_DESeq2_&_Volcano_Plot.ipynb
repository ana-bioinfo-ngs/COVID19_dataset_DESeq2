DESeq2

```python
# Install and Load PyDESeq2
%pip install pydeseq2
```

```python
from pydeseq2.dds import DeseqDataSet #importing class DeseqDataSet from the dds module of pydeseq2 package
from pydeseq2.ds import DeseqStats  #importing class DeseqStats from the ds module of pydeseq2 package

import pandas as pd
```

```python
counts = pd.read_csv('/content/Combined_RAW_Counts.csv')
counts
```

```python
counts = counts.set_index('gene_id')
counts
```

```python
counts = counts[counts.sum(axis = 1) > 0] # selects and keeps only the rows whose sum across all samples is > 0
counts
```

```python
counts = counts.T # Transpose the columns to rows as this is required for the next pydeseq2 tools
counts
```

```python
data = zip(counts.index, ['Infected','Infected','Infected', 'Mock', 'Mock', 'Mock']) # combines each element of index column list and new conditions list
metadata = pd.DataFrame(data, columns = ['Sample', 'Condition']) # creates a table for sample data and condition
```

```python
metadata = metadata.set_index('Sample') # sets Sample column as index
metadata
```

```python
dds = DeseqDataSet(
    counts=counts,
    metadata=metadata,
    design_factors="Condition"
) # All the data is being loaded in DeseqDataset class
```

```python
dds.deseq2() #Performs Differential Expression Analysis
```

```python
dds # Result is stored in dds object
```

```python
stat_res = DeseqStats(dds, n_cpus=8, contrast = ('Condition','Infected','Mock')) # computes differential expression statistics by comparing Infected against Mock samples
```

```python
stat_res.summary() # gives a readable summary of the analysis
```

```python
res = stat_res.results_df # gives tabular dataframe of the results
res
```

```python
res.to_csv("DESeq2_results_I_vs_M.csv") # save results in csv file
```

Volcano Plot

```python
from sanbomics.plots import volcano
```

```python
res
```

```python
volcano(res, symbol='Symbol') # plotting volcano plot using symbols column
```

```python
logfc_cutoff = 1        # |log2FC| ≥ 1
padj_cutoff = 0.05
```

```python
upregulated = res[
    (res['log2FoldChange'] >= logfc_cutoff) &
    (res['padj'] < padj_cutoff)
]
```

```python
downregulated = res[
    (res['log2FoldChange'] <= -logfc_cutoff) &
    (res['padj'] < padj_cutoff)
]
```

```python
print("Upregulated genes:", upregulated.shape[0])
print("Downregulated genes:", downregulated.shape[0])
```
